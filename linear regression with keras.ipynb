```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/DL0101EN/labs/data/concrete_data.csv')
```

```python
data_columns = data.columns
```

```python
predictors = data[data_columns[data_columns != 'Strength']]
```

```python
predictors.head()
```

```python
target = data['Strength']
```

```python
pre_norms = (predictors - predictors.mean()/predictors.std())
```

```python
pre_norms.head()
```

```python
n_cols = pre_norms.shape[1]
```

```python
import keras
from keras.models import Sequential
from keras.layers import Dense
```

```python
def regression():
    model = Sequential()
    model.add(Dense(50, activation='relu', input_shape=(n_cols,)))
    model.add(Dense(50, activation='relu'))
    model.add(Dense(1))
    
    
    model.compile(optimizer ='adam', loss='mean_squared_error')
    return model
    
```

```python
model = regression()
```

```python
model.fit(pre_norms,target, validation_split=.3, epochs=100, verbose=2)
```
